```python
# Convert the date column to datetime

operations["date"] = pd.to_datetime(
    operations["date"],
    format="mixed"
)
```

```python
operations.info()
```

```python
# Check numeric columns that should contain positive values

positive_columns = [
    "trade_volume_tonnes",
    "freight_cost_usd",
    "carbon_emissions_tonnes"
]

(operations[positive_columns] <= 0).sum()
```
