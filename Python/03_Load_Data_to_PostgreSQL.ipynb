# Load Cleaned Data into PostgreSQL

This notebook connects Python to PostgreSQL using SQLAlchemy and loads the cleaned datasets into the database for SQL analysis and Power BI reporting.

```python
import pandas as pd

from sqlalchemy import create_engine
from urllib.parse import quote_plus
```

```python
username = "postgres"
password = quote_plus("Vasa@2004")

host = "localhost"
port = "5432"

database = "supply_chain_analysis"

engine = create_engine(
    f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"
)

with engine.connect():
    print("Connection Successful!")
```

```python
country_metadata = pd.read_csv("cleaned_country_metadata.csv")

trade_routes = pd.read_csv("cleaned_trade_routes.csv")

weekly_timeline = pd.read_csv("cleaned_weekly_timeline.csv")

commodity_market = pd.read_csv("cleaned_commodity_market.csv")

geopolitical_events = pd.read_csv("cleaned_geopolitical_events.csv")

operations = pd.read_csv("cleaned_weekly_route_operations.csv")
```

```python
country_metadata.to_sql(
    "country_metadata",
    engine,
    if_exists="replace",
    index=False
)

trade_routes.to_sql(
    "trade_routes",
    engine,
    if_exists="replace",
    index=False
)

weekly_timeline.to_sql(
    "weekly_timeline",
    engine,
    if_exists="replace",
    index=False
)

commodity_market.to_sql(
    "commodity_market",
    engine,
    if_exists="replace",
    index=False
)

geopolitical_events.to_sql(
    "geopolitical_events",
    engine,
    if_exists="replace",
    index=False
)

operations.to_sql(
    "operations",
    engine,
    if_exists="replace",
    index=False
)

print("All datasets loaded successfully!")
```
